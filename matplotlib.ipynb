Matplotlib
- useful for making Plots

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for the plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
#sine wave
plt.plot(x,y) #here x is the axis and y=sin(x) represented on the y-axis
plt.show()
```

```python
#build a cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# plotting a parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+') # r represents the red color and we plot it using the '+' symbol
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #here 0,0 is the coordinates of the origin, and 1,1 are the rectangle sides length
languages = ['arabic','english','spanish','german']
people = [100,50,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['arabic','english','spanish']
people = [50,30,20]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D scatter plot

```python
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
