Pandas dataframe is two-dimensional data structure with labeled axes (rows and columns)


```python
#importing the pandas library
import pandas as pd
import numpy as np
```

creating a pandas dataframe

```python
#importing the california housing dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
type(housing)
```

```python
print(housing)
```

```python
#pandas Dataframe
housing_df = pd.DataFrame(housing.data,columns = housing.feature_names)
```

Manipulating a dataframe


to add a column we write :

---

df['new_col_name'] = data

---
and to remove and row/column we write:


---

df.drop(index=x,axis=y) where x is the index of the row or column we want to remove and y = 0, if it's a row else 1 if it's a column.


```python
#locate particular row using index value
housing_df.iloc[2]
```

```python
# locate a particular column
print(housing_df.iloc[:,0]) #first column
```
