Methods to Handle missing values :
* imputation
* dropping

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset to a pandas dataframe
dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

we have 67 missing values in salary column

Central Tendencies:


1.   mean
2.   median
3.   mode



```python
# analyse the distributions
#fig,ax = plt.subplots(figsize=(8,8))
sns.displot(dataset['salary'])
```

because of outliers, we cannot use the mean, because the mean is sensitive towards outliers.

Replace the missing values with Median values

```python
dataset['salary'].fillna(dataset['salary'].median(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset['salary']
```

Dropping Method

```python
# we create another dataset to test the dropping method
salary_dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
salary_dataset.isnull().sum()
```

```python
#drop missing values
salary_dataset = salary_dataset.dropna(how='any')
```
